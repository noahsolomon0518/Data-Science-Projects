# Goal: 
The goal of this notebook is to classify whether a mushroom is edible or poisonous using a neural network.The data set that I am using is:
https://archive.ics.uci.edu/ml/datasets/Mushroom

Mushrooms will be classfied into 2 categories, either poisonous or edible. There are 22 features, which categorically describe the aspects of each mushroom.  More information on the features can be found on the UCI link.

## Section 1: Cleaning the Data

```python
#Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mushrooms.data')
df = df.drop(['e.1'], axis = 1)       #Dropping column with majority of empty data
enc = LabelEncoder()
for i in range(len(df.columns)):      #For each column
    enc.fit(df.iloc[:,i])
    df.iloc[:,i] = enc.transform(df.iloc[:,i])
df
```

### Data Exploration

```python
feature = 'n'
for x in range(len(df.columns)):
    feature = df.columns[x]
    print('--------------------------------------------------')
    print('TESTING FOR:', feature)
    print()
    for i in range(len(df[feature].unique())):   
        val = i
        spec = df[df.x == val]
        pois = spec[spec.p==1]
        edib = spec[spec.p==0]
        print("When", feature, '=', val, 'there are:')
        print(len(pois),'poisonous')
        print(len(edib),'edible')
```

All features seem relevant so none will be removed. 

## Section 2: Creation of Neural Network

```python
#Create the model
model = Sequential()
model.add(Dense(1000,input_dim = 21, activation = 'relu'))
model.add(Dense(1,activation = 'sigmoid'))
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

## Prepare Data

```python
x = df.iloc[:,1:]
y = df.p
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = .2)
```

## Testing

```python
model.fit(x_train,y_train, epochs = 30)
```

```python
model.evaluate(x_test, y_test)      #100% accuracy!
```

```python
pred = model.predict(x_test)
```

```python
#Compare predicted with actual 
predList = []
actList = []
for i in pred:
    if i>.50:
        predList.append('Poisonous')
    else:
        predList.append('Edible')

        
for i in y_test:
    if i==1:
        actList.append('Poisonous')
    else:
        actList.append('Edible')



for i in range(len(predList)):
    print('Predicted: ', predList[i], '||   Actual: ' , actList[i])
```

## Conclusion
This data has well defined relationships between all the features and labels allowing a neural network to easily find patterns. Now given you find a mushroom and wonder if it is poisonous, its features can be inputed into the neural network and it can make a highly educated guess to whether it is safe to eat.
